## Calculate the angle of the EPC building location relative to the Google Street View camera location

```python
import pandas as pd

dt = pd.read_excel('d321b25ef1a30bb1.xlsx')
dt = dt[['WALLS_DESC',
                             'lat', 'lng', 'folder',
                             'name','FLOOR_LEVE','lat_1','lon']]
dt
type(dt)
```

```python
dt.WALLS_DESC.unique()
```

```python
dt['angle'] = (dt.lng.astype(float) - dt.lon.astype(float))/(dt.lat.astype(float) - dt.lat_1.astype(float))
```

```python
dt
```

```python
import math
import numpy as np
dt['angle'] = np.degrees(np.arctan(dt['angle'])) 
dt
```

```python
dt['lat1'] = np.radians(dt.lat_1)
dt['lat2']= np.radians(dt.lat)
dt['diffLong'] = np.radians(dt.lng - dt.lon)
dt['x'] = np.sin(dt.diffLong) * np.cos(dt.lat2)
dt['y'] = np.cos(dt.lat1) * np.sin(dt.lat2) - (np.sin(dt.lat1)
            * np.cos(dt.lat2) * np.cos(dt.diffLong))
dt['initial_bearing'] = np.arctan2(dt.x, dt.y)
dt['initial_bearing'] = np.degrees(dt.initial_bearing)
dt['angle'] = (dt.initial_bearing + 360) % 360
dt
```

```python
final = dt[['WALLS_DESC','folder','name', 'angle','FLOOR_LEVE','lat', 'lng', 'lat_1','lon']]
```
